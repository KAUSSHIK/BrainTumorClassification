#DATA PREPARATION

```python
labels = ['glioma_tumor','no_tumor','meningioma_tumor','pituitary_tumor']

#We start off by appending all the images from the directories into a list and then converting it into a NumPy array after resizing the images to 150x150.

#X_train is a list of all the images in the training set.
X_train = []
#y_train is a list of all the labels in the training set.
y_train = []
#image_size is the size of the images in the dataset.
image_size = 150

#cv2.imread() is a function in OpenCV that reads an image from a file.
#cv2.resize() is a function in OpenCV that resizes an image.
for i in labels:
    folderPath = './data/Training/' + i + '/'
    for filename in os.listdir(folderPath):
        img = cv2.imread(folderPath + filename)
        img = cv2.resize(img, (image_size, image_size))
        X_train.append(img)
        y_train.append(labels.index(i))

for i in labels:
    folderPath = './data/Testing/' + i + '/'
    for filename in os.listdir(folderPath):
        img = cv2.imread(folderPath + filename)
        img = cv2.resize(img, (image_size, image_size))
        X_train.append(img)
        y_train.append(labels.index(i))

X_train = np.array(X_train)
y_train = np.array(y_train)
print(X_train.shape)
```

Why Do We Need One-hot Encoding?
When working with categorical variables like the types of tumors in this classification model, it's crucial to convert these categories into a numerical format that can be ingested by machine learning algorithms. One-hot encoding is one such technique for this transformation.

Problem with Label Encoding
If we merely label the four tumor types as 0, 1, 2, and 3, the machine learning algorithm might misinterpret these numbers as having ordinal significance. For example, the model might incorrectly assume that type 3 is "greater than" or "three times" the type 0, which doesn't make logical sense in this context.

The One-hot Encoding Solution
In one-hot encoding, each categorical label is transformed into a binary vector. For example:

0 (Glioma tumor) becomes [1, 0, 0, 0]
1 (No tumor) becomes [0, 1, 0, 0]
2 (Meningioma tumor) becomes [0, 0, 1, 0]
3 (Pituitary tumor) becomes [0, 0, 0, 1]
Each position in the vector corresponds to a category, and a 1 is placed in the index that matches the category label, leaving all other positions as 0.

Advantages
No Arbitrary Ranking: With one-hot encoding, we avoid implying any ordinal relationship between categories.

Independence: Each category is represented independently, allowing the model to learn the unique features of each class.

Better Performance: Models often perform better with one-hot encoded data than with label-encoded data, particularly when there's no ordinal relationship between the categories.

Therefore, one-hot encoding is an essential preprocessing step to ensure the best possible performance of our machine learning model.

```python
#One hot encoding the labels
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

## What is Transfer Learning?

Transfer learning is a technique in machine learning where a model trained on one task is adapted for a second related task. In the realm of deep learning, and especially for tasks like image classification, transfer learning can substantially improve both the training efficiency and the performance of your model.

### Why is Transfer Learning Important?

1. **Time and Computational Efficiency**: Training a deep neural network from scratch can be computationally expensive and time-consuming. Transfer learning speeds up this process by reusing parts of an already trained model.

2. **Improved Performance**: Pre-trained models like EfficientNet have been trained on large datasets like ImageNet, which has over 14 million labeled images. Leveraging these pre-trained models provides a robust feature extractor that can capture relevant features from images.

3. **Data Scarcity**: When you have a limited amount of labeled data for your specific problem, transfer learning comes in handy. The pre-trained model serves as a general feature extractor that can be fine-tuned for your specific classification task.

### Typical Workflow in Transfer Learning

1. **Feature Extraction**: The pre-trained model serves as a feature extractor. When an image is fed into this model, it generates a set of "features" representing the content of the image.

2. **Fine-Tuning**: These features are then fed into a new neural network (usually consisting of fully connected layers), which is specifically trained to perform the task at hand—in this case, brain tumor classification.

By employing transfer learning, we can achieve excellent classification performance with less data and reduced training time.


```python
effnet = EfficientNetB0(weights='imagenet',include_top=False,input_shape=(image_size,image_size,3))

# EfficientNetB0: This is a specific architecture among EfficientNet models, known for their efficiency and performance.

# weights='imagenet': Specifies that the model should use the weights pre-trained on the ImageNet dataset.

# include_top=False: This means the final fully connected layer for classification in the pre-trained model will be omitted, allowing us to add custom layers suited for our problem.

# input_shape=(image_size,image_size,3): This defines the shape of the input image as a 3D tensor with dimensions for height, width, and RGB color channels.
```

## Simplified Explanation of Neural Network Layers

### GlobalAveragePooling2D
Think of this layer as a way to simplify the information coming from the preceding layers. If the previous layer outputs a bunch of numbers, this layer takes the average of those numbers. This has two benefits:

1. **Data Reduction**: It makes the incoming data smaller and easier to manage, speeding up the network.
2. **Feature Retention**: It retains the essential features necessary for classification.

### Dropout
When you're training a neural network, sometimes the model starts "memorizing" the training data instead of "learning" from it. This is bad because the model will not generalize well to new, unseen data. To avoid this, the Dropout layer randomly "turns off" a fraction of neurons during each training step. By "turning off," we mean those neurons do not update during that particular training step. This makes it hard for the network to memorize the training data, thus helping the model generalize better.

### Dense (Output Layer)
The Dense layer here is your "decision-maker." After all the simplifications and feature extractions from the above layers, this layer decides which class a given input most likely belongs to. Since there are 4 different types of tumors, this layer has 4 units. The softmax activation function assigns probabilities to these 4 classes, and the class with the highest probability is the model's final output.


```python
model = effnet.output
model = keras.layers.GlobalAveragePooling2D()(model) #takes the average of the numbers coming from the previous layer to simply computation
model = keras.layers.Dropout(rate=0.5)(model) #randoly drops 50% of the neurons
model = keras.layers.Dense(4, activation='softmax')(model) #output layer with 4 units and softmax activation function
model = keras.models.Model(inputs=effnet.input, outputs=model) #creates the model
```

### Understanding model.compile()

When you're just starting out with machine learning and deep learning, terms like "loss function," "optimizer," and "metrics" can be a bit overwhelming. But don't worry, let's break down what each term means in simple language.

#### What is it doing?

The `model.compile()` method is like setting the rules of the game for our neural network. Imagine you're teaching a child how to play a new game. You'll explain how to score points (metrics), what counts as a mistake (loss), and how to improve (optimizer).

##### Key Components:

1. **Loss Function (`categorical_crossentropy`)**: Think of this as the "mistake meter." It's a way to measure how far off our model's predictions are from the actual results. The goal of training a model is to make this number as low as possible by adjusting the internal settings (or "parameters").
    
2. **Optimizer (`Adam`)**: This is the strategy the model will use to minimize the "mistakes" (loss). Just like a coach in a game would advise players on techniques to improve their gameplay, the optimizer advises the model on how to update itself to make fewer mistakes.
    
3. **Metrics (`accuracy`)**: This is how we keep score. The accuracy tells us the proportion of predictions our model gets right. So, if the accuracy is 0.90 or 90%, it means our model makes the correct prediction 90% of the time.

By running `model.compile()`, we're telling the model: "Hey, here's how we measure mistakes, here's how you should correct them, and here's how we'll keep track of your performance."

---




```python
model.compile(loss='categorical_crossentropy',optimizer = 'Adam', metrics= ['accuracy'])
#metrics = ['accuracy'] tells the model to keep track of the accuracy score during training. Will judge based on how ACCRUATE the model is
```

### Understanding Callbacks: TensorBoard, ModelCheckpoint, and ReduceLROnPlateau

When you're new to Machine Learning (ML) and Deep Learning (DL), the term "callbacks" may sound strange. Simply put, callbacks are like coaches or trainers who monitor the model during training and intervene when necessary. They can help you see what's happening during training, save your model, and even alter your learning rate for better performance. Let's dive into the specifics.

#### What are these Callbacks Doing?

1. **TensorBoard (`tensorboard`)**: Imagine you are playing a video game and you have a dashboard that shows your score, health, and other important statistics. TensorBoard is like that dashboard but for your machine learning model. It allows you to visualize various metrics like loss, accuracy and even the architecture of your model in real-time as it trains.
    - `log_dir = 'logs'`: This is where TensorBoard will save its log files. You can use these log files to visualize the training process later.

2. **ModelCheckpoint (`checkpoint`)**: You wouldn't want to lose your high score in a game due to a power cut, right? Similarly, ModelCheckpoint saves the "best" version of your model as it trains, so you don't lose it.
    - `monitor="val_accuracy"`: It keeps an eye on the model's validation accuracy.
    - `save_best_only=True`: Saves only the "best" (highest validation accuracy) model.
    - `mode="auto",verbose=1`: Automatically decides whether to maximize or minimize the monitored metric, and prints a message whenever it saves a new "best" model.

3. **ReduceLROnPlateau (`reduce_lr`)**: Think of this as a smart coach that helps the model when it's stuck. If the model isn't improving, this callback reduces the learning rate to help the model get out of the plateau (a point where the model isn't improving).
    - `monitor = 'val_accuracy'`: Watches the model's validation accuracy.
    - `factor = 0.3`: Reduces the learning rate to 30% of the current learning rate.
    - `patience = 2`: Waits for 2 epochs before reducing the learning rate.
    - `min_delta = 0.001`: Requires a minimum change in validation accuracy to count as an improvement.
    - `mode='auto',verbose=1`: Automatically decides whether accuracy should be maximized or minimized and prints a message whenever it reduces the learning rate.

Remember, each of these callbacks has a special role, and using them in combination can make your model training far more effective and easier to understand.

---

```python
tensorboard = TensorBoard(log_dir = 'logs')
checkpoint = ModelCheckpoint("effnet.h5",monitor="val_accuracy",save_best_only=True,mode="auto",verbose=1) 
#effnet.h5 is where the checkpoint will be saved
reduce_lr = ReduceLROnPlateau(monitor = 'val_accuracy', factor = 0.3, patience = 2, min_delta = 0.001,mode='auto',verbose=1)
```

## Training the Model with `model.fit`

Now that we've prepared our data and compiled our model, it's finally time to train! Let's understand the key elements of the `model.fit()` function.

### Key Parameters:

- **X_train, y_train**: These are the variables that contain our training data and corresponding labels.

- **validation_split**: We've set this to 0.1, meaning that 10% of our training data will be set aside to validate the model's performance.

- **epochs**: This is the number of times the model will run through the entire training dataset. We've set it to 12.

- **verbose**: This parameter controls what gets printed during training. Setting it to 1 will print a detailed log.

- **batch_size**: The model updates its internal parameters not based on each sample, but after every 32 samples, which is the batch size.

- **callbacks**: These are extra functionalities we've added for monitoring and improving the model's performance. 
  - `tensorboard` for visualizing metrics.
  - `checkpoint` for saving the best-performing model.
  - `reduce_lr` for automatically reducing the learning rate when the model's performance plateaus.

### What Happens During Training?

1. The model starts its training cycle using `X_train` and `y_train`.
2. After processing each batch of 32 samples, the model updates its internal parameters to minimize the loss function.
3. Once an epoch is complete, the model uses the validation set to evaluate its performance.
4. The best-performing model parameters are saved, thanks to our `checkpoint` callback.
5. If the model's performance isn't improving, the `reduce_lr` callback automatically adjusts the learning rate.

This way, we aim to train a model that generalizes well to new, unseen data.


## Understanding Training, Validation, and Test Sets

When we're training a machine learning model, it's essential to understand the difference between training, validation, and test sets.

### Training Set
The **training set** is what the model learns from. It adjusts its internal settings or parameters based on the input data and the corresponding labels in the training set.

### Validation Set
The **validation set** is a separate portion of the data used during training but not for learning. After each round (or "epoch") of learning on the training set, the model's performance is evaluated using the validation set. This evaluation helps us tune hyperparameters and decide when to stop training. In our case, we set `validation_split=0.1` to use 10% of the training data as a validation set.

### Test Set
The **test set** is used after the model is fully trained. It provides an unbiased evaluation of the model's performance on completely new data that it has never seen. This helps us gauge how well the model will perform when deployed in the real world.

### Why Separate Sets?
Having separate validation and test sets allows for a more comprehensive evaluation. The validation set helps during the model-building process, letting us make real-time adjustments. The test set gives us a final, unbiased measure of our fully trained model's performance.

By using all three sets properly, we can more confidently assess the effectiveness and robustness of our machine learning model.


```python
history = model.fit(X_train,y_train,validation_split=0.1, epochs =12, verbose=1, batch_size=32,
                    callbacks=[tensorboard,checkpoint,reduce_lr])
```

```python
# Get the history dictionary from the history object
history_dict = history.history

# Extract the training and validation accuracy
train_acc = history_dict['accuracy']
val_acc = history_dict['val_accuracy']

#Etract the training and validation loss
train_loss = history_dict['loss']
val_loss = history_dict['val_loss']

# Get the number of epochs
epochs = range(1, len(train_acc) + 1)

# Create a new figure
plt.figure(figsize=(12, 6))

#Subplots
plt.subplot(1, 2, 1) #1 row, 2 columns, 1st subplot
plt.plot(epochs, train_acc, 'bo-', label='Training Accuracy')
plt.plot(epochs, val_acc, 'ro-', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Subplot for Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, train_loss, 'bo-', label='Training Loss')
plt.plot(epochs, val_loss, 'ro-', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

## Model Prediction and Evaluation

After training the model, the next crucial step is to evaluate its performance on unseen data, also known as the test set. The code snippet provided accomplishes this task as follows:

1. **Making Predictions**: `pred = model.predict(X_test)`
    - This line uses the trained model to make predictions on the test set (`X_test`). The model returns an array of probabilities for each of the classes.

2. **Converting Probabilities to Class Labels**: `pred = np.argmax(pred, axis=1)`
    - The model's output is in the form of probabilities. To convert these into actual class labels, we use `np.argmax` to find the index (class label) with the maximum probability for each sample. The `axis=1` argument specifies that we are looking for the maximum value across each row (i.e., for each sample).

3. **Preparing Ground Truth Labels**: `y_test_new = np.argmax(y_test, axis=1)`
    - The actual labels (`y_test`) are in one-hot encoded format. To make them directly comparable with our predictions, we convert them back into integer format using `np.argmax`.

After these steps, both the predicted labels (`pred`) and actual labels (`y_test_new`) will be in integer format, making it easy to evaluate the model's performance using various metrics like accuracy, precision, recall, etc.

```python
pred = model.predict(X_test)
pred = np.argmax(pred,axis=1)
y_test_new = np.argmax(y_test,axis=1)
```

## Understanding the Classification Report

After running our model on the test data, we generate a detailed performance report using `classification_report` from the scikit-learn library. This report contains several key metrics that help us understand how well our model is performing:

### Key Metrics:

1. **Precision**: Measures the accuracy of the positive predictions. It tells us the ratio of true positive predictions to the total number of positive predictions (true positives + false positives).
  
2. **Recall**: Also known as sensitivity, recall tells us the ratio of true positive predictions to the total actual positives (true positives + false negatives). It helps us understand how many of the actual positive cases our model is able to capture.

3. **F1-score**: This is the harmonic mean of Precision and Recall and gives us a balanced measure of the two metrics. An F1-score closer to 1 indicates a better model.

4. **Support**: This shows the number of occurrences of each class in the ground truth (y_test_new).

5. **Accuracy**: This tells us how often the classifier is correct across all classes.

6. **Macro avg**: This is the average of the above metrics without considering the class distribution, giving us an equally weighted average.

7. **Weighted avg**: This gives us the average considering the class distribution (i.e., it's a weighted average).

By understanding these metrics, we can get a comprehensive view of how well our model performs, not just in terms of overall accuracy but also in how well it identifies each class.

```python
# Classification Report
print(classification_report(y_test_new,pred))
```

```python
from ipywidgets import widgets, FileUpload, Button, Output, VBox
import io
from PIL import Image
import cv2
import numpy as np
from tensorflow import keras  # Importing the keras library from TensorFlow
import keras

model = keras.models.load_model('effnet.h5')


def img_pred(upload, model):
    upload_tuple = upload.value
    if isinstance(upload_tuple, tuple) and len(upload_tuple) > 0:
        upload_dict = upload_tuple[0]
        if 'content' in upload_dict:
            img_content = upload_dict['content']
            img = Image.open(io.BytesIO(img_content))
            opencvImage = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)
            img = cv2.resize(opencvImage, (150, 150))
            img = img.reshape(1, 150, 150, 3)
            p = model.predict(img)
            p = np.argmax(p, axis=1)[0]
            
            if p == 0:
                p = 'Glioma Tumor'
            elif p == 1:
                print('The model predicts that there is no tumor')
            elif p == 2:
                p = 'Meningioma Tumor'
            else:
                p = 'Pituitary Tumor'
            
            if p != 1:
                print(f'The Model predicts that it is a {p}')
        else:
            print("Content not found in upload")
    else:
        print("Unexpected data structure:", upload_tuple)




uploader = FileUpload()
display(uploader)

button = Button(description='Predict')
output = Output()  # Use the same name here

def on_button_clicked(_):
    with output:  # And also here
        output.clear_output()  # Use the correct output object
        print(f"Debug: uploader.value = {uploader.value}")
        try:
            img_pred(uploader, model)  # Pass the model as a second argument
        except Exception as e:
            print('No Image Uploaded/Invalid Image File:', str(e))
        
button.on_click(on_button_clicked)
display(VBox([button, output]))
```
